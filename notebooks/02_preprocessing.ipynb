```python
import torch
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
from medmnist import ChestMNIST
import numpy as np
import matplotlib.pyplot as plt

# Seed fixe pour la reproductibilité
torch.manual_seed(42)
np.random.seed(42)

print("PyTorch version :", torch.__version__)
print("Seed fixé à 42")
```

```python
MEAN = 0.5057
STD  = 0.2896

train_transform = transforms.Compose([
    transforms.Resize((64, 64)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[MEAN], std=[STD]),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(degrees=10),
])

test_transform = transforms.Compose([
    transforms.Resize((64, 64)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[MEAN], std=[STD]),
])

print("Transformations définies avec taille 64x64")
```

```python
train_dataset = ChestMNIST(split='train', transform=train_transform, download=True, size=64)
val_dataset   = ChestMNIST(split='val',   transform=test_transform,  download=True, size=64)
test_dataset  = ChestMNIST(split='test',  transform=test_transform,  download=True, size=64)

print(f"Train : {len(train_dataset)} images")
print(f"Val   : {len(val_dataset)} images")
print(f"Test  : {len(test_dataset)} images")
```
